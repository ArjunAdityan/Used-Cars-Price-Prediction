```python
import re

# Define a function to extract the value before 'Cylinder'
def extract_cylinder_info(text):
    match = re.search(r'(\b(?:Straight\s*\d+|V\d+|\d+))\s+Cylinder', text, re.IGNORECASE)
    if match:
        return match.group(1)
    else:
        return '-1'

# Apply the function to create a new column
df['Cylinder Info'] = df['engine'].apply(extract_cylinder_info)
```

```python
# Define a function to extract the phrase containing 'engine'
def extract_engine_phrase(text):
    match = re.search(r'(\b\w+\b\s*Engine)', text, re.IGNORECASE)
    if match:
        return match.group(1)
    else:
        return '-1'

df['engine type'] = df['engine'].apply(extract_engine_phrase)
```

```python
# Define a function to extract the text after 'V'
def extract_after_v(text):
    match = re.search(r'V\s*(.*)', text)
    if match:
        return match.group(1)
    else:
        return '-1'

# Apply the function to create a new column
df['dummy'] = df['engine'].apply(extract_after_v)
df.loc[df['dummy'] != '-1', 'engine type'] = df['dummy']

# Drop the helper column
df = df.drop(columns=['dummy'])
```

```python
df.loc[df['engine'].str.contains('Electric Motor'), 'engine type'] = 'Electric Motor'
```

```python
# Define a function to extract the phrase starting with 'Liter'
def extract_from_liter(text):
    match = re.search(r'Liter.*', text)
    if match:
        return match.group(0)
    else:
        return '-1'

# Apply the function and update the existing column conditionally
df['dummy'] = df['engine'].apply(extract_from_liter)
df.loc[df['dummy'] != '-1', 'engine type'] = df['dummy']

# Drop the helper column
df = df.drop(columns=['dummy'])
```

```python
# Define a function to remove the first character if it is numeric
def remove_first_numeric(text):
    if text and text[0].isdigit():
        return text[1:].strip()
    return text

# Apply the function to update the existing column
df['engine type'] = df['engine type'].apply(remove_first_numeric)
```

```python
# Define the function to remove the first word if the value starts with a numeric character
def remove_first_word_if_numeric(value):
    if value[0].isdigit():
        return ' '.join(value.split(' ')[1:])
    else:
        return value

# Apply the function to the 'engine_spec' column
df['engine type'] = df['engine type'].apply(remove_first_word_if_numeric)
```

```python
# Define the function to extract voltage
def extract_voltage(value):
    match = re.search(r'(\d+)(?=V)', value)
    if match:
        return int(match.group(1))
    else:
        return -1

# Apply the function to create a new column
df['voltage'] = df['engine'].apply(extract_voltage)
```
